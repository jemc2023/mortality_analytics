```python
target_schema = "semi2.sandbox"
accion = "DROP" 

try:
    tables_df = spark.sql(f"SHOW TABLES IN {target_schema}")
    tables_list = [row.tableName for row in tables_df.collect()]
    
    if not tables_list:
        print(f"El esquema '{target_schema}' ya está completamente vacío.")
    else:
        for table_name in tables_list:
            full_path = f"{target_schema}.{table_name}"
            
            if accion == "TRUNCATE":
                spark.sql(f"TRUNCATE TABLE {full_path}")
            elif accion == "DROP":
                spark.sql(f"DROP TABLE IF EXISTS {full_path}")
                
        print(f"\nOperación finalizada con éxito sobre el esquema '{target_schema}'.")
except Exception as e:
    print(f"Error al procesar las tablas: {e}")
```
